# **Listas**

```python
#Criando uma lista chama animais
animais = [1, 2, 3]
animais
```

```python
animais = ["cachorro", "gato", 12345, 6.5]
animais
```

```python
# Imprimindo o primeiro elemento da lista
animais[0]
```

```python
# Imprimindo o quarto elemento da lista
animais[3]
```

```python
# Substituindo o primeiro elemento da lista
animais[0] = "papagaio"
animais
```

```python
# Removendo gato da lista
animais.remove("gato")
animais
```

```python
# Tamanho da lista
len(animais)
```

```python
# Verificando se um elemento esta na lista
"gato" in animais
```

```python
lista = [500, 30, 300, 80, 10]
```

```python
# Verificando o maior valor
max(lista)
```

```python
# Verificando o menor valor
min(lista)
```

```python
# Adicionando um elemento a lista
animais.append("leão")
animais
```

```python
# Adicionando mais de um elemento a lista
animais.extend(["cobra", 6])
animais
```

```python
animais.count("cachorro")
```

```python
# Ordenando a lista
lista.sort()
lista
```

# **Tuplas**

```python
# As tuplas usam parentese como sintaxe
tp = ("Banana", "Maçã", 10, 50)
```

```python
# Retornando o primeiro elemento
tp[0]
```

```python
# Diferente das listas, tuplas são imutáveis, o que quer dizer que não podemos alterar seus elementos
tp[0] = "Laranja"
```

```python
tp.count("Maçã")
```

```python
tp[0:2]
```

# **Dicionários**

```python
# Para criar um dicionário utilizamos as {}
dc = {"Maçã" : 20, "Banana" : 10, "Laranja" : 15, "Uva" : 5}
# Dicionários trabalham com o conceito de chave e valor
dc
```

```python
# Acessando o valor de um dicionário através da chave
dc["Maçã"]
```

```python
# Atualizando o valor da maçã
dc["Maçã"] = 25
dc["Maçã"]
```

```python
# Retornando todas as chaves do dicionário
dc.keys()
```

```python
# Retornando todos os valores do dicionário
dc.values()
```

```python
# Verificando se já existe uma chave no dicionário e caso não exista inserir
dc.setdefault("Limão", 22)
dc
```
